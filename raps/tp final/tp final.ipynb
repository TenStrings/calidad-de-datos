```r
library(tidyverse)
library(arules)
```

```r
df<-read.csv("Covid19Casos.csv", fileEncoding = "UTF-8")
```

```r
df
```

```r

datos<-filter(df, !is.na(fecha_inicio_sintomas) & as.Date(fecha_inicio_sintomas,"%Y-%m-%d")>=as.Date("2020-05-01") & as.Date(fecha_inicio_sintomas,"%Y-%m-%d")<=as.Date("2020-05-31"))
```

```r
mayo<-datos %>% select(sexo,edad,residencia_provincia_nombre,fecha_inicio_sintomas,fallecido,clasificacion_resumen,asistencia_respiratoria_mecanica,cuidado_intensivo)
```

```r
cambiar_calor<-function( col,  val,  replace){
    mayo$col <- as.character(mayo$col)

    mayo$col[mayo$col == val] <- string replace

    mayo$col <- as.factor(mayo$col)
    
}
```

```r
mayo$fallecido <- as.character(mayo$fallecido)

mayo$fallecido[mayo$fallecido == "NO"] <- "NO_fallecido"

mayo$fallecido <- as.factor(mayo$fallecido)
#-------------------------------------------------
mayo$fallecido <- as.character(mayo$fallecido)

mayo$fallecido[mayo$fallecido == "SI"] <- "SI_fallecido"

mayo$fallecido <- as.factor(mayo$fallecido)
#-----------------------------------------------------
mayo$asistencia_respiratoria_mecanica <- as.character(mayo$asistencia_respiratoria_mecanica)

mayo$asistencia_respiratoria_mecanica[mayo$asistencia_respiratoria_mecanica == "NO"] <- "NO_respirador"

mayo$asistencia_respiratoria_mecanica <- as.factor(mayo$asistencia_respiratoria_mecanica)

mayo$asistencia_respiratoria_mecanica <- as.character(mayo$asistencia_respiratoria_mecanica)

mayo$asistencia_respiratoria_mecanica[mayo$asistencia_respiratoria_mecanica == "SI"] <- "SI_respirador"

mayo$asistencia_respiratoria_mecanica <- as.factor(mayo$asistencia_respiratoria_mecanica)
#------------------------------------------------------------------------------------
mayo$cuidado_intensivo <- as.character(mayo$cuidado_intensivo)

mayo$cuidado_intensivo[mayo$cuidado_intensivo == "NO"] <- "NO_intensivo"

mayo$cuidado_intensivo <- as.factor(mayo$cuidado_intensivo)

mayo$cuidado_intensivo <- as.character(mayo$cuidado_intensivo)

mayo$cuidado_intensivo[mayo$cuidado_intensivo == "SI"] <- "SI_intensivo"

mayo$cuidado_intensivo <- as.factor(mayo$cuidado_intensivo)
#------------------------------------------------------------------------------------

```
