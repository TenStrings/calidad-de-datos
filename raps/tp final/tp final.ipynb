```r
library(tidyverse)
library(arules)
```

```r
df<-read.csv("Covid19Casos.csv", fileEncoding = "UTF-8")
```

```r

datos<-filter(df, !is.na(fecha_inicio_sintomas) & as.Date(fecha_inicio_sintomas,"%Y-%m-%d")>=as.Date("2020-05-01") & as.Date(fecha_inicio_sintomas,"%Y-%m-%d")<=as.Date("2020-05-31"))
```

```r

datosjunio<-filter(df, !is.na(fecha_inicio_sintomas) & as.Date(fecha_inicio_sintomas,"%Y-%m-%d")>=as.Date("2020-06-01") & as.Date(fecha_inicio_sintomas,"%Y-%m-%d")<=as.Date("2020-05-30"))
```

```r

txs <- read.transactions("mayo_edades.csv", format = "basket", sep = ',', cols=1)
```

```r
reglas <- apriori(txs, parameter=list(support=0.1,confidence = 0.7))

test <- arules::subset(reglas, subset = lhs %in% c("Confirmado"))

inspect(sort(test, by="lift"))
```
